## Собираем входящие сообщение от пользователя
Создадим копию датафрейма `df_cm`

```python
df = df_cm.copy()
df[(df['created_by'] == 0) & (df['type'] == 'outgoing_chat_message')]
```

Мы обнаружили 173 сообщения, которые были помечены как отправленные менеджером, но имели идентификатор 0. Это явный сбой в работе системы.

Поскольку в условиях чётко указано, что идентификатор 0 всегда присваивается клиенту, я принял решение изменить тип этих 173 строк на «входящий», чтобы они соответствовали ситуации, когда клиент отправляет сообщение менеджеру.

```python
df.type.value_counts()
```

```python
df['type'] = df.apply(lambda x: 'incoming_chat_message' if (x.created_by == 0) & (x.type == 'outgoing_chat_message') else x.type, axis=1)
df.type.value_counts()
```

Переводим время сообщение на нормальное вместо `Unix Timestamp`

```python
df.loc[:, ['created_at']] = df['created_at'].apply(lambda x: pd.Timestamp(x, unit='s', tz='Europe/Moscow'))
df.head()
```

Для более удобного восприятия данных, мы рекомендуем отсортировать их по `entity` и `created_at`.

```python
df = df.sort_values(by=['entity_id', 'created_at'])
df.head(10)
```

Теперь нужно добавить дополнительные столбцы: Время и тип сообщения на предыдущей строке.

```python
df['previous_type'] = df.groupby('entity_id')['type'].shift()
df['previous_time'] = df.groupby('entity_id')['created_at'].shift()
df.head(10)
```

Теперь, когда мы создали необходимые столбцы, у нас есть возможность определить новый признак, основанный на времени отправки каждого ответа.

При этом важно, чтобы значения создавались только для исходящих сообщений, а тип сообщения не повторялся в двух и более строках подряд.

Это позволит нам учесть ситуации, когда один и тот же адресат отправляет несколько сообщений подряд. В конце процедуры мы удалим ненужные столбцы.

```python
df['question_time'] = df.apply(lambda x: x['previous_time'] if (x['type'] != x['previous_type']) & (x['type'] == 'outgoing_chat_message') else None, axis=1)
df = df.drop(['previous_type', 'previous_time'], axis=1)
df.head(10)
```

Далее показываю исходящие ответы менеджеров, которые были отправлены с `0:00` до `9:30`

```python
df[(df['created_at'].dt.time.between(pd.Timestamp('00:00:00').time(), pd.Timestamp('09:30:00').time())) & (df['type'] == 'outgoing_chat_message') & (df['question_time'].notna())]
```

В нашей выборке есть случаи, когда менеджер приходил на работу раньше и сразу отвечал на сообщения, а также когда он задерживался допоздна и отвечал на них уже после полуночи. Иногда менеджер отвечал на сообщения даже ночью.

Поскольку разница во времени ответов в этих случаях была очень большой, а расчёт среднего значения всегда чувствителен к аномальным значениям, я решил исключить их из нашей выборки.

Кроме того, мы удаляем все входящие сообщения и исходящие, для которых не было вопроса от клиента. Сделка же представляет собой одно исходящее сообщение.

```python
df = df[(df['question_time'].notna()) & (df['type'] == 'outgoing_chat_message') & (~df['created_at'].dt.time.between(pd.Timestamp('00:00:00').time(), pd.Timestamp('09:30:00').time()))]
```

Далее определим функцию расчета разницы между временем вопроса и временем ответа.

1. Определем дату и время вопроса`q_time` и ответа `a_time`
2. Определяем переменную `work_time_start`, которая будет определять начало рабочего дня и будем изменять дату в соответствии с датой ответа
3. Для случая, когда вопрос был получен с полуночи до начала рабочего дня, разница будет определяться между временем ответа и временем начала рабочего дня
4. Для случаев, когда вопрос был получен до полуночи, а ответ был отправлен утром после начала рабочего дня, разница будет считаться между временем ответа и вопроса
5. Во всех остальных случаях мы просто получаем разницу между временем ответа и временем вопроса

```python
def calculate_time_diff(row):
    q_time = row['question_time']
    a_time = row['created_at']

    work_time_start = a_time.normalize() + pd.Timedelta(hours=9, minutes=30)

    if (q_time >= a_time.normalize()) and (q_time < work_time_start) and (a_time >= work_time_start):
        return a_time - work_time_start
    elif q_time < a_time.normalize() and a_time >= work_time_start:
        return a_time - q_time - pd.Timedelta(hours=9, minutes=30)
    else:
        return a_time - q_time
```

```python
df['time_diff'] = df.apply(calculate_time_diff, axis=1)
df.head(20)
```

Теперь переведем полученное значения разницы в минуты для лучшего восприятия.

Для этого сначала переведем тип в формат секунд, возьмем их целое кол-во и разделим на 60

```python
df['time_diff_min'] = round(df['time_diff'].astype('timedelta64[s]').astype('int64') / 60, 2)
df.head(20)
```

Теперь можем рассчитать среднее кол-во минут на ответ каждого менеджера.

Группировку производим по столбцу `created_by`.

```python
df['avg_time_diff_min'] = round(df.groupby('created_by')['time_diff_min'].transform('mean'))
df.head(20)
```

Теперь скомпануем итоговую таблицу.

Сначала соединим текущую таблицу с датасетом, содержащим информацию о менеджерах. Выберем только нужные столбцы и уберем дубликаты.

```python
df_final = pd.merge(df, df_m, left_on='created_by', right_on='mop_id', how='right')[['name_mop', 'avg_time_diff_min']].drop_duplicates().sort_values('avg_time_diff_min')
df_final
```
